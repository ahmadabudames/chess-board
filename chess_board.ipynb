```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
class chess_board:
      def __init__(self):
        self.grid = np.ones((8,8,3),np.float64)
        self.arr_red = []
        self.arr_blue = []
        div = 1
        for col in range(8):
            for row in range(8):
                div = 0 if div else 1 
                if row % 2 == div and col % 2 ==div:
                    self.grid[col][row] =np.array([0,0,0])
      def add_red(self ,col , row):
            self.grid[col][row] =np.array([1,0,0])
            self.arr_red=[col , row ]
            print("red " ,self.arr_red)
      def add_blue(self ,col , row):
            self.grid[col][row] = np.array([0,0,1])
            self.arr_blue=[col , row ]
            print("bule arr " ,self.arr_blue)
      def is_under_attack(self, red, blue):
        x = abs(red[0] - blue[0])
        y = abs(red[1] - blue[1])
        if (x == 0) or  (y == 0) or (x == y):
            return True
        else:
            return False
      def render(self):
         plt.imshow(self.grid)


checkboard= chess_board()
checkboard.add_blue(6,4)
checkboard.add_red(6,3)   
checkboard.render() 

# test queens on same row should be “under attack”

assert checkboard.is_under_attack([6,4],[6,3])==True
```

```python

checkboard= chess_board()
checkboard.add_blue(0,3)
checkboard.add_red(6,3)   
checkboard.render() 

# test queens on same column should be “under attack”

assert checkboard.is_under_attack([0,3],[6,3])==True
```

```python
checkboard= chess_board()
checkboard.add_blue(4,5)
checkboard.add_red(6,3)   
checkboard.render() 

# test queens on same diagonal should be “under attack”

assert checkboard.is_under_attack([4,5],[6,3])==True
```

```python
checkboard= chess_board()
checkboard.add_blue(5,1)
checkboard.add_red(6,3)   
checkboard.render() 

# test queens with any other coordinates should NOT be “under attack”

assert checkboard.is_under_attack([5,1],[6,3])==False
```
